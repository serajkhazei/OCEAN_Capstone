# Creating master dataframe

# Libaries

```python
import pandas as pd
```

```python
df = pd.read_csv('mypersonality_cleaned.csv')
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
df.drop(['#AUTHID','sentiNEG','sentiPOS','sentiNEU','sentiCOMPOUND','DATE',
         'NETWORKSIZE','BETWEENNESS','NBETWEENNESS','DENSITY',
         'BROKERAGE','NBROKERAGE','TRANSITIVITY','cEXT','cNEU','cAGR','cCON','cOPN'],axis=1, inplace=True)
```

```python
df
```

```python
df.rename(columns={'sEXT': "Extraversion", 'sNEU': "Neuroticism", 'sAGR':"Agreeablenes",
                   'sCON':"Consciousness" ,'sOPN':"Openness",'STATUS':"status_update"},inplace=True)
```

```python
df = df.reindex(['status_update','Openness','Consciousness','Extraversion','Agreeablenes','Neuroticism'], axis=1)
```

```python
df
```

```python
df2 = pd.read_csv('final_df.csv')
```

```python
df2.drop('userid',axis=1, inplace=True)
```

```python
df2.rename(columns={'sEXT': "Extraversion", 'sNEU': "Neuroticism", 
                   'sAGR':"Agreeablenes",'sCON':"Consciousness" ,'sOPN':"Openness"},inplace=True)
```

```python
df2
```

```python
master_df = pd.concat([df, df2],ignore_index=True)
master_df
```

```python
master_df.to_csv('master_data.csv')
```
